```python
pre_survey = survey[['Gender', 'Age', 'Do you have thoughts of suicide?', 'What are you concerned about today?', 
               'On a scale of 1-5, how upset are you?']]
```

```python
pre_survey = pre_survey.dropna()
pre_survey.head()
```

```python
from sklearn.feature_extraction import DictVectorizer
categorical_features = pre_survey[['Gender', 'Age', 'On a scale of 1-5, how upset are you?', 'Do you have thoughts of suicide?',
                                   'What are you concerned about today?']]

dv = DictVectorizer()
cat_matrix = dv.fit_transform(categorical_features.T.to_dict().values())
```

```python
type(cat_matrix)
cat_matrix.todense() 
```

```python
#from scipy.sparse import hstack
#other_features = survey[['Age']]
#data_matrix = hstack([cat_matrix, other_features])
data_matrix = cat_matrix
data_matrix
```

```python
from sklearn.preprocessing import scale

data_matrix = scale(data_matrix.todense())
```

```python
from sklearn.cluster import KMeans

clustering_model = KMeans(n_clusters = 25)
clustering_model.fit(data_matrix)

clusters = clustering_model.predict(data_matrix)

clusters
```

```python
results = pd.DataFrame({ 'cluster' : clusters, 
                        'What are you concerned about today?' : pre_survey['What are you concerned about today?'],
                        'Do you have thoughts of suicide?': pre_survey['Do you have thoughts of suicide?'],
                        'Gender': pre_survey['Gender'],
                        'On a scale of 1-5, how upset are you?': pre_survey['On a scale of 1-5, how upset are you?']
                        })
print results.head()
#cluster_counts = results.groupby('cluster').agg(len) #.value_counts() # DataFrame Groupby ['What are you concerned about today?']

#cluster_counts = results.groupby('cluster')['What are you concerned about today?'].value_counts()
#print cluster_clounts
cluster_counts = results.groupby('cluster').agg(len)
```
